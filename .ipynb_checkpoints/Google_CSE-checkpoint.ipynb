```python
import gensim
from gensim import corpora, models, similarities
import logging
logging.basicConfig(format='%(asctime)s : %(levelname)s : %(message)s', level=logging.INFO)


def tokenize(content):
    """
    Tokenize a piece of text from wikipedia. The input string `content` is assumed
    to be mark-up free (see `filter_wiki()`).
  
    Return list of tokens as utf8 bytestrings. Ignore words shorted than 2 or longer
    that 15 characters (not bytes!).
    """
    # TODO maybe ignore tokens with non-latin characters? (no chinese, arabic, russian etc.)
    return [token.encode('utf8') for token in utils.tokenize(content, lower=True, errors='ignore')
            if 2 <= len(token) <= 15 and not token.startswith('_')]

texts = [tokenize(r) for r in raw_texts]
```

```python
stoplist = set('as such they are am this that those this these is if when for a of the and to in'.split())
dictionary = corpora.Dictionary(texts)
# remove stop words and words that appear only once
stop_ids = [dictionary.token2id[stopword] for stopword in stoplist if stopword in dictionary.token2id]
once_ids = [tokenid for tokenid, docfreq in dictionary.dfs.iteritems() if docfreq == 1]
dictionary.filter_tokens(stop_ids + once_ids) # remove stop words and words that appear only once
dictionary.compactify() # remove gaps in id sequence after words that were removed
print(dictionary)
# dictionary.token2id
```

```python
def cleanText(corpura):
    for i in range(len(corpura)):
        corpus=corpura[i].encode('utf-8')
        
        punctuation = """#@?!.,:;(){}[]$1234567890\/-+%"""

        corpus = corpus.lower().replace('\n','')
        corpus = corpus.replace('<br />', ' ')
        #treat punctuation as individual words
        
        for c in punctuation:
            corpus = corpus.replace(c,"",10)
#         stoplist = set('about it is by be being that this those these donald trump I my or other mine me you your yours he him his she her we are our ours yours their theirs they it has have on had as such they are at from with am this that those this these is if when for of the and to in'.split())
        
        corpus = corpus.split()
        corpus = [z for z in corpus if len(z)>1]
        corpura[i]=corpus
    return corpura

news_text = cleanText(news_text)
```

```python
def buildDocHistogram(Vocab_ind, text, ind_size,normalize='Yes'):
    vec = np.zeros(ind_size).reshape((1, ind_size))
    count = 0.
    for word in text:
        try:
            vec[0,Vocab_ind[word]] += 1
            count += 1.
        except KeyError:
            continue
    if count != 0:
        if normalize=='Yes':
            vec /= count
    return vec
```

```python
#Build new dim for vocabs based on SOMinds
print 'data size', VocabVec.shape
ind_final_vocab = VocabVec.sum(axis=1)!=0

final_VocabVec = VocabVec[ind_final_vocab]
print "final_VocabVec size ", final_VocabVec.shape
final_vocab = list(np.asarray(vocab)[ind_final_vocab])
Vocab_Wordind = dict(zip(final_vocab,range(len(final_vocab)) ))
```

```python
Googlevec = 'No'
#Build word vector for training set by using the average value of all word vectors in the tweet, then scale
def buildWordVector(text,model_w2v, size):
    vec = np.zeros(size).reshape((1, size))
    count = 0.
    for word in text:
        try:
            vec += model_w2v[word].reshape((1, size))
            count += 1.
        except KeyError:
            continue
    if count != 0:
        vec /= count
    return vec


from sklearn.preprocessing import scale


if Googlevec=='Yes':
    train_vecs = np.concatenate([buildWordVector(z, Google_w2v,n_dim) for z in x_train])
    train_vecs = scale(train_vecs)
    test_vecs = np.concatenate([buildWordVector(z,Google_w2v, n_dim) for z in x_test])
    test_vecs = scale(test_vecs)
else:
    train_vecs = np.concatenate([buildWordVector(z, imdb_w2v,n_dim) for z in x_train])
    train_vecs = scale(train_vecs)
    test_vecs = np.concatenate([buildWordVector(z,imdb_w2v, n_dim) for z in x_test])
    test_vecs = scale(test_vecs)


# #Train word2vec on test tweets
# imdb_w2v.train(x_test)
#Build test tweet vectors then scale

```

```python
def buildDocHistogram(Vocab_ind, text, ind_size,normalize='Yes'):
    vec = np.zeros(ind_size).reshape((1, ind_size))
    count = 0.
    for word in text:
        try:
            vec[0,Vocab_ind[word]] += 1
            count += 1.
        except KeyError:
            continue
    if count != 0:
        if normalize=='Yes':
            vec /= count
    return vec
```

```python
#now select the most informative features (here are sominds, but we can do this on original words too)

def calc_conditional_feature_importance(corpus_mat,labels):
    #corpus_mat is the original matrix where each row is one record and columns are features, where are either words or sominds
    #sentiments are labels
    #it returns a matrix showing the relative importance of each feature regarding to each label
    pos_coocur_ = np.zeros((corpus_mat.shape[1],1))
    neg_coocur_ = np.zeros((corpus_mat.shape[1],1))
    ind_pos = labels==1
    ind_neg = labels==0
    for i in range(corpus_mat.shape[1]):
        pos_coocur_[i] = np.sum(corpus_mat[ind_pos,i])
        neg_coocur_[i] = np.sum(corpus_mat[ind_neg,i])
        sum_ = (pos_coocur_[i]+neg_coocur_[i])
        pos_coocur_[i] = pos_coocur_[i]/sum_
        neg_coocur_[i] = neg_coocur_[i]/sum_
    DF =pd.DataFrame(index=range(corpus_mat.shape[1]))
    DF['pos_coocur_'] = pos_coocur_
    DF['neg_coocur_'] = neg_coocur_
    DF['differ'] = np.abs(neg_coocur_-pos_coocur_)
    DF = DF.sort(columns='differ',ascending=False)
    return DF
```

```python
def bootstrap_predict(X_train,y_train,X_test,y_test,method='SOM',n_estimators=10,max_features=1,max_sample=1,
                      bootsrap_feature=False,msz=[30,30],knn=1):
    # X_train = np.asarray([[1,2,3,4],[5,6,7,8]])
    MAE = pd.DataFrame(data=np.zeros((1,0)))
    y_preds = pd.DataFrame(data=np.zeros((y_test.shape[0],0)))
    ARE = pd.DataFrame(data=np.zeros((y_test.shape[0],0)))
#     n_estimators = 30
#     max_features = 1
#     max_sample = 1
#     bootsrap_feature = False
    dlen = X_train.shape[0]
    dim = X_train.shape[1]
    for i in range(n_estimators):
        sample_i_index = np.random.random_integers(0,high=dlen-1,size=int(max_sample*dlen))
        if bootsrap_feature == True:
            feature_i_index = np.random.random_integers(0,high=dim-1,size=int(max_features*dim))
            sample_i = X_train[sample_i_index][:,feature_i_index]
            X_test_i = X_test[:,feature_i_index]
        
#         std = sample_i.std(axis=0)
#         randomization = 1e-35*np.random.randn(sample_i.shape[0],1)
#         ind_std0 = std==0
#         sample_i[:,ind_std0] = sample_i[:,ind_std0] + randomization 
        
        
        
        elif bootsrap_feature == False:
            length =  X_train.shape[1]
            ind = np.asarray([ np.random.randint(0,2) for x in xrange(length) ])
            sample_i = X_train[sample_i_index][:,ind>0]
            X_test_i = X_test[:,ind>0]
        
        
#         std = sample_i.std(axis=0)
#         randomization = 1e-35*np.random.randn(sample_i.shape[0],1)
#         ind_std0 = std==0
#         sample_i[:,ind_std0] = sample_i[:,ind_std0] + randomization 
        
        
        y_train_i = y_train[sample_i_index]
    
#         msz = [100,100]
#         knn = 1
#         method = 'SOM'
        preds = predict_one_step(sample_i,X_test_i,y_train_i,knn=knn,which=method,msz=msz).values[:,0]
    
        y_preds[str(i)+'_msz_'+str(msz[0])+'_k_'+str(knn)] = preds
        MAE[str(i)+'_msz_'+str(msz[0])+'_k_'+str(knn)] = np.mean(np.abs(preds-y_test))
        ARE[str(i)+'_msz_'+str(msz[0])+'_k_'+str(knn)] = 100*np.abs(preds-y_test)/y_test

    ensemble_mean = y_preds.mean(axis=1)
    ensemble_median = y_preds.median(axis=1)
    y_preds['ensemble_mean'] = ensemble_mean
    MAE['ensemble_mean'] = np.mean(np.abs(ensemble_mean-y_test))
    ARE['ensemble_mean'] = 100*np.abs(ensemble_mean-y_test)/y_test

    y_preds['ensemble_median'] = ensemble_median
    MAE['ensemble_median'] = np.mean(np.abs(ensemble_median-y_test))
    ARE['ensemble_median'] = 100*np.abs(ensemble_median-y_test)/y_test
    
    return y_preds,ARE,MAE

# method='RndForest'
# preds = predict_one_step(X_train,X_test,y_train,knn=knn,which=method,msz=msz).values[:,0]
# y_preds[method] = preds
# MAE[method] = np.mean(np.abs(preds-y_test))
# ARE[method] = 100*np.abs(preds-y_test)/y_test

# method='SOM'
# preds = predict_one_step(X_train,X_test,y_train,knn=knn,which=method,msz=msz).values[:,0]
# y_preds[method] = preds
# MAE[method] = np.mean(np.abs(preds-y_test))
# ARE[method] = 100*np.abs(preds-y_test)/y_test

# method='lin'
# preds = predict_one_step(X_train,X_test,y_train,knn=knn,which=method,msz=msz).values[:,0]
# y_preds[method] = preds
# MAE[method] = np.mean(np.abs(preds-y_test))
# ARE[method] = 100*np.abs(preds-y_test)/y_test

# MAE[['ensemble','ensemble_median','RndForest','SOM','lin']]
```

```python
def predict_one_step(X_train,X_test,y_train,knn=5,which='lin',msz=[30,40]):
    
    from sklearn import linear_model
    from sklearn.svm import SVR

    y_preds = pd.DataFrame(data=np.zeros((X_test.shape[0],1)))
#     which = 'Ridge'
    approximation_method = which
    if approximation_method=='lin':
        regr = linear_model.LinearRegression()
        name = 'lin'
        regr.fit(X_train, y_train)
        y_preds.ix[:,0:1]=regr.predict(X_test)[:,np.newaxis]

    elif approximation_method=='Ridge':
#         regr = linear_model.Ridge(alpha=.1)
        regr = linear_model.RidgeCV(alphas=[0.1, .5,1.0,2.,5, 10.0])
        name = 'Ridge'
        regr.fit(X_train, y_train)
        y_preds.ix[:,0:1]=regr.predict(X_test)[:,np.newaxis]

    elif approximation_method=='Bayes_Ridge':
        regr = linear_model.BayesianRidge()
        name = 'Bayes_Ridge'            
        regr.fit(X_train, y_train)
        y_preds.ix[:,0:1]=regr.predict(X_test)[:,np.newaxis]
    
    
    elif approximation_method=='SVR_RBF':
        regr = SVR(kernel='rbf', C=1e3, gamma=0.1)
        name = 'SVR_RBF'
        regr.fit(X_train, y_train)
        y_preds.ix[:,0:1]=regr.predict(X_test)[:,np.newaxis]

    elif approximation_method=='SVR_Lin':
        regr = SVR(kernel='linear', C=1e3)
        name = 'SVR_Lin'
        regr.fit(X_train, y_train)
        y_preds.ix[:,0:1]=regr.predict(X_test)[:,np.newaxis]

    elif approximation_method=='SVR_poly':
        regr = SVR(kernel='poly', C=1e3, degree=1)
        name = 'SVR_poly'
        regr.fit(X_train, y_train)
        y_preds.ix[:,0:1]=regr.predict(X_test)[:,np.newaxis]

    elif approximation_method=='LassoLars':
        regr = linear_model.LassoLars(alpha=.1)
        name = 'LassoLars'
        regr.fit(X_train, y_train)
        y_preds.ix[:,0:1]=regr.predict(X_test)[:,np.newaxis]

    elif approximation_method=='ARDR':
        regr = linear_model.ARDRegression()
        name = 'ARDR'
        regr.fit(X_train, y_train)
        y_preds.ix[:,0:1]=regr.predict(X_test)[:,np.newaxis]

    elif approximation_method=='ElasticNet':
        regr = linear_model.ElasticNet()
        name = 'ElasticNet'
        regr.fit(X_train, y_train)
        y_preds.ix[:,0:1]=regr.predict(X_test)[:,np.newaxis]

    elif approximation_method=='LogisticRegression':
        regr = linear_model.LogisticRegression()
        name = 'LogisticRegression'
        regr.fit(X_train, y_train)
        y_preds.ix[:,0:1]=regr.predict(X_test)[:,np.newaxis]

    elif approximation_method=='Lasso':
        regr = linear_model.Lasso()
        name = 'Lasso'
        regr.fit(X_train, y_train)
        y_preds.ix[:,0:1]=regr.predict(X_test)[:,np.newaxis]

    elif approximation_method=='OMP':
        regr = linear_model.OrthogonalMatchingPursuit()
        name = 'OMP'
        regr.fit(X_train, y_train)
        y_preds.ix[:,0:1]=regr.predict(X_test)[:,np.newaxis]

    elif approximation_method=='Perceptron':
        regr = linear_model.Perceptron()
        name = 'Perceptron'
        regr.fit(X_train, y_train)
        y_preds.ix[:,0:1]=regr.predict(X_test)[:,np.newaxis]

    elif approximation_method=='SGD':
        regr = linear_model.SGDRegressor()
        name = 'stochastic_gradient'
        regr.fit(X_train, y_train)
        y_preds.ix[:,0:1]=regr.predict(X_test)[:,np.newaxis]
    elif approximation_method=='knn':
        from sklearn import neighbors
        regr = neighbors.KNeighborsRegressor(knn, weights = 'distance')
        regr.fit(X_train, y_train)
        y_preds.ix[:,0:1]=regr.predict(X_test)[:,np.newaxis]
    elif approximation_method=='knn_dis':
        from sklearn import neighbors
        regr = neighbors.KNeighborsRegressor(knn, weights = 'distance')
        regr.fit(X_train, y_train)
        y_preds.ix[:,0:1]=regr.predict(X_test)[:,np.newaxis]
    
    elif approximation_method == 'RndForest':
        import sklearn.ensemble as ensemble
        regr = ensemble.RandomForestRegressor(n_jobs=1) 
        regr.fit(X_train, y_train)
        y_preds.ix[:,0:1]=regr.predict(X_test)[:,np.newaxis]
    
    elif approximation_method == 'AdaBoost':
        import sklearn.ensemble as ensemble
        regr = ensemble.AdaBoostRegressor() 
        regr.fit(X_train, y_train)
        y_preds.ix[:,0:1]=regr.predict(X_test)[:,np.newaxis]
        
    elif approximation_method == 'Bagging':
        import sklearn.ensemble as ensemble
        regr = ensemble.BaggingRegressor(n_jobs=1)
        regr.fit(X_train, y_train)
        y_preds.ix[:,0:1]=regr.predict(X_test)[:,np.newaxis]
    elif approximation_method == 'D-tree':
        from sklearn.tree import DecisionTreeRegressor
        # Fit regression model
        regr = DecisionTreeRegressor()
        # Predict
        regr.fit(X_train, y_train)
        y_preds.ix[:,0:1]=regr.predict(X_test)[:,np.newaxis]
        
    elif which == 'SOM':
        All_Data_train = np.concatenate((X_train,y_train[:,np.newaxis]),axis=1)    
        reload(sys.modules['sompy'])
        somP = SOM.SOM('som', All_Data_train, mapsize = msz,norm_method = 'var',initmethod='pca')
        somP.train(n_job = 1, shared_memory = 'no',verbose='off')
        
        y_preds.ix[:,0:1] = somP.predict(X_test,K=knn)[:,np.newaxis]
        
    ##Prediction

#     print 'done'
    return y_preds
```
